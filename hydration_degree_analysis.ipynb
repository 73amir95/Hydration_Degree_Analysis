```python
def add_percentage_diff_columns(df):
    for col in df.columns:
        col_diffs = []
        for j in range(len(df) - 1):
            old_value = df.iloc[j][col]
            new_value = df.iloc[j + 1][col]
            diff = (new_value - old_value) / old_value * 100
            col_diffs.append(diff)
        # Append NaN to keep the length consistent
        col_diffs.append(None)
        # Create the new column name
        new_col_name = f"{col}_diff [%]"
        # Insert the new column into the DataFrame
        df[new_col_name] = col_diffs
```

```python
add_percentage_diff_columns(df_mean)
df_mean
```

```python
add_percentage_diff_columns(df_std)
df_std
```
